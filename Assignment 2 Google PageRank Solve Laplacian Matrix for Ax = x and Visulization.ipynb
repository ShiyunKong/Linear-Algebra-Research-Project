```python

# Your provided data
matches = [
    ('tr', 'it', 0, 1), ('wa', 'ch', 0, 0), ('dk', 'fi', 0, 1),
    ('be', 'ru', 1, 0), ('en', 'hr', 1, 0), ('at', 'ma', 1, 0),
    ('nl', 'ua', 1, 0), ('sc', 'cz', 0, 1), ('pl', 'sk', 0, 1),
    ('sp', 'se', 0, 0), ('hu', 'pt', 0, 1), ('fr', 'de', 1, 0),
    ('fi', 'ru', 0, 1), ('tr', 'wa', 0, 1), ('it', 'ch', 1, 0),
    ('ua', 'ma', 1, 0), ('dk', 'be', 0, 1), ('nl', 'at', 1, 0),
    ('se', 'sk', 1, 0), ('hr', 'cz', 0, 0), ('en', 'sc', 0, 0),
    ('hu', 'fr', 0, 0), ('pt', 'de', 0, 1), ('sp', 'pl', 0, 0),
    ('it', 'wa', 1, 0), ('ch', 'tr', 1, 0), ('ma', 'nl', 0, 1),
    ('ua', 'at', 0, 1), ('ru', 'dk', 0, 1), ('fi', 'be', 0, 1),
    ('cz', 'en', 0, 1), ('hr', 'sc', 1, 0), ('sk', 'sp', 0, 1),
    ('se', 'pl', 1, 0), ('de', 'hu', 0, 0), ('pt', 'fr', 0, 0),
    ('wa', 'dk', 0, 1), ('it', 'at', 1, 0), ('nl', 'cz', 0, 1),
    ('be', 'pt', 1, 0), ('hr', 'sp', 0, 1), ('fr', 'ch', 0, 0),
    ('en', 'de', 1, 0), ('se', 'ua', 0, 1), ('ch', 'sp', 0, 0),
    ('it', 'be', 1, 0), ('cz', 'dk', 0, 1), ('en', 'ua', 1, 0),
    ('it', 'sp', 0, 0), ('en', 'dk', 1, 0), ('it', 'en', 0, 0),
    # ('sk', 'it', 1, 0)
]

# Create a set of unique team names
teams = set(match[0] for match in matches) | set(match[1] for match in matches)

# Create an empty adjacency matrix
adjacency_matrix = np.zeros((len(teams), len(teams)))

# Populate the adjacency matrix with scores
for match in matches:
    team1, team2, score1, score2 = match
    index1 = list(teams).index(team1)
    index2 = list(teams).index(team2)
    adjacency_matrix[index1, index2] = score1
    adjacency_matrix[index2, index1] = score2

print(adjacency_matrix)
```

```python
A = np.dot(p, adjacency_matrix) + np.dot((1 - p) * 1/n, np.ones((24, 24)))
A
```

```python
# method of normalize
def normalize_columns(matrix):
    c_max = 1
    column_sums = np.sum(matrix, axis=0)  # Calculate the sum of each column

    # Avoid division by zero
    column_sums[column_sums == 0] = 1

    # Normalize the columns
    normalized_matrix = matrix / column_sums * c_max

    return normalized_matrix
A = normalize_columns(A)
A
```

```python
# method of finding Pagerank
def pagerank(connectivity_matrix):
    n = connectivity_matrix.shape[0]

    # Initialize the PageRank vector
    v = np.zeros((n, 1))
    v[1] = 1

    # matrix_A = connectivity_matrix * 0.85 + 0.15 * (1/n) * np.ones(n)

    # Calculate Pagerank
    epochs = 2**10
    for i in range(epochs):
        v = np.dot(connectivity_matrix , v)
    return v
pageranks = pagerank(A)
```

```python
rank = pd.DataFrame({
    'name': teams,
    'rank': [pageranks.tolist()[i][0] for i in range(len(pageranks))]
}).sort_values('rank',ascending=False).reset_index()
rank
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# Create an empty graph
G = nx.DiGraph()
plt.figure(figsize=(10, 10))
plt.suptitle("European Championship 2021 Matches", size=24, weight='bold')
plt.title("Arrow points from winner to loser or points in both direction for tie", size=18, weight='bold')
# Add nodes and edges based on your Google PageRank matrix
for match in matches:
    team1, team2, score1, score2 = match
    if score1 < score2 :
        G.add_edge(team2, team1)
    elif score1 > score2 :
        G.add_edge(team1, team2)
    elif score1 == score2:
        G.add_edge(team2, team1)
        G.add_edge(team1, team2)
# Draw the graph with node sizes proportional to PageRank scores
nx.draw(G, with_labels=True, node_size=[int(rank[rank['name'] == node]['rank'].tolist()[0] * 10000) for node in list(G.nodes())], pos=nx.kamada_kawai_layout(G), node_color='orange', font_color='b', font_size=20)
plt.show()
```
