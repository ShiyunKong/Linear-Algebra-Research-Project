```python
import numpy as np

# Define the range and number of points
start_value = -2
end_value = 2
num_points = 101

# Generate equidistant points
equidistant_points = np.linspace(start_value, end_value, num_points)

# Print the result
print(equidistant_points)
```

```python
import numpy as np

# Generate equidistant points
t = np.linspace(start_value, end_value, 101)

# Generate normally distributed random variables
mean = 0
std_dev = 5
random_variables = np.random.normal(mean, std_dev, 101)

# Calculate yi = 4t^3 - 4t + random_variable
yi = 4 * t**3 - 4 * t + random_variables

# Print the result
print(yi)
```

```python
import numpy as np

# Generate equidistant points
t = np.linspace(0, 1, 101)

# Create the Vandermonde matrix
A = np.vander(t, increasing=True)

# Include additional columns for t_i, t_2i, and t_3i
A[:, 1] = t
A[:, 2] = t**2
A[:, 3] = t**3

# Print the result
A = A[:, :4]
A
```

```python
import numpy as np

# Assuming A and y are already defined
# If not, generate A using the Vandermonde matrix code from the previous responses
# Also, generate y using the formula 4t^3 - 4t + random_variable

# Solving the least squares problem
x, residuals, rank, singular_values = np.linalg.lstsq(A, yi, rcond=None)

# Print the result
print("Solution vector x:", x)
print("Residuals:", residuals)
print("Rank of the coefficient matrix A:", rank)
print("Singular values of A:", singular_values)
```

```python
# Calculate Mean Squared Error
from sklearn.metrics import r2_score
mse = np.mean((yi - (x[-1] * t**3 + x[-2] * t**2 + x[-3] * t + x[-4]))**2)
mse2 = np.mean((yi - (xx[-1] * t**2 + xx[-2] * t + xx[-3]))**2)
mse3 = np.mean((yi - (xxx[-1] * t + xxx[-2]))**2)
r_squared = r2_score(yi, (x[-1] * t**3 + x[-2] * t**2 + x[-3] * t + x[-4]))
r_squared2 = r2_score(yi, (xx[-1] * t**2 + xx[-2] * t + xx[-3]))
r_squared3 = r2_score(yi, (xxx[-1] * t + xxx[-2]))

print("Mean Squared Error of Cubic fitting:", mse)
print("R-squared of Cubic fitting:", r_squared)
print("Mean Squared Error of Quadratic fitting:", mse2)
print("R-squared of Quadratic fitting:", r_squared2)
print("Mean Squared Error of Linear fitting:", mse3)
print("R-squared of Linear fitting:", r_squared3)
```

```python
# Calculate residuals
residuals_3 = yi - (x[-1] * t**3 + x[-2] * t**2 + x[-3] * t + x[-4])
residuals_2 = yi - (xx[-1] * t**2 + xx[-2] * t + xx[-3])
residuals_1 = yi - (xxx[-1] * t + xxx[-2])
# Create a residual plot using Seaborn and Matplotlib
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x[-1] * t**3 + x[-2] * t**2 + x[-3] * t + x[-4], y=residuals_3, color='r', alpha=1, label='Cubic')
sns.scatterplot(x=xx[-1] * t**2 + xx[-2] * t + xx[-3], y=residuals_2, color='green', alpha=1, label='Quadratic')
sns.scatterplot(x=xxx[-1] * t + xxx[-2], y=residuals_1, color='orange', alpha=1, label='Linear')
plt.axhline(y=0, color='b', linestyle='--', linewidth=2)  # Reference line at y=0
plt.title('Residual Plot')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```
